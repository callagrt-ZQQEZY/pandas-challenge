```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
csv_path = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(csv_path, encoding="utf-8")

purchase_data_df.head()

```

## Player Count

* Display the total number of players


```python
players = purchase_data_df ["SN"].nunique()

players_df = pd.DataFrame({"Number of Players": [players]})
players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
 



items_count = len(purchase_data_df["Item Name"].unique())
average_price = purchase_data_df["Price"].mean()
purchase_count = len(purchase_data_df["Purchase ID"].unique())

total_revenue = purchase_data_df["Price"].sum()

summary_df = pd.DataFrame({"Number of Unique Items": [items_count], 
                           "Average Price": average_price,
                           "Number of Purchases": purchase_count, 
                           "Total Revenue": total_revenue})
summary_df["Average Price"] = summary_df["Average Price"].astype(float).map("${:,.2f}".format)

summary_df["Total Revenue"] = summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
purchase_data_df.dtypes
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_stats = purchase_data_df.groupby("Gender")

gender_stats_df = pd.DataFrame(gender_stats[["SN", "Purchase ID",]].nunique())
gender_stats_df.head()
```

```python

gender_stats_df = gender_stats_df.rename(
    columns={"SN": "Players", "Purchase ID": "Purchase Count"})
gender_stats_df.head()
```

```python
gender_prices = purchase_data_df.groupby("Gender")

gender_prices_value = pd.DataFrame(gender_prices["Price"].sum())
gender_prices_value.head()
```

```python
gender_prices_value = gender_prices_value.rename(
    columns={"Price": "Total Purchase Value"})
gender_prices_value.head()
```

```python
merged_df = gender_stats_df.merge(gender_prices_value, 
                                  on=["Gender"])

merged_df.head()
```

```python
merged_df["Average Purchase Price"] = merged_df["Total Purchase Value"] /\
                                                    merged_df["Purchase Count"]
merged_df["Avg Total Purchase per Person"] = merged_df["Total Purchase Value"] /\
                                                    merged_df["Players"] 

merged_df["Average Purchase Price"] = merged_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)

merged_df["Avg Total Purchase per Person"] = merged_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

merged_df["Total Purchase Value"] = merged_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)

merged_df
```

```python
reduced_merged = pd.DataFrame(merged_df[["Purchase Count",
                                          "Average Purchase Price",
                                          "Total Purchase Value",
                                          "Avg Total Purchase per Person"]])
reduced_merged
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
Age_Demo_df = pd.DataFrame(purchase_data_df[["SN",
                                          "Age"]])
Age_Demo_df.head()
```

```python
Age_Demo_df.describe()
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]

group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
Age_Demo_df["Age Groups"] = pd.cut(Age_Demo_df["Age"], bins, labels=group_names, include_lowest=True)
Age_Demo_df
```

```python
age_group = Age_Demo_df.groupby("Age Groups")
age_analysis_df = pd.DataFrame(age_group["SN"].nunique())
age_analysis_df.head(10)
```

```python
age_analysis_df = age_analysis_df.rename(
    columns={"SN": "Total Count"})
age_analysis_df.head(10)
```

```python
age_analysis_df["Percentage of Players"] = (age_analysis_df["Total Count"]/age_analysis_df["Total Count"].sum())*100


age_analysis_df["Percentage of Players"] = age_analysis_df["Percentage of Players"].astype(float).map("{:,.2f}%".format) 

age_analysis_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purchases_df = pd.DataFrame(purchase_data_df[["SN", "Purchase ID", "Price",
                                          "Age"]])
age_purchases_df.head()
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]

group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
age_purchases_df["Age Ranges"] = pd.cut(age_purchases_df["Age"], bins, labels=group_names, include_lowest=True)
age_purchases_df
```

```python
age_purchase_count = age_purchases_df.groupby("Age Ranges")
age_grouping_df = pd.DataFrame(age_purchase_count[["SN", "Purchase ID",]].nunique())
age_grouping_df.head(10)
```

```python
age_grouping_df = age_grouping_df.rename(
    columns={"SN": "Players", "Purchase ID": "Purchase Count"})
age_grouping_df.head(10)
```

```python
age_prices = age_purchases_df.groupby("Age Ranges")

age_prices_value = pd.DataFrame(age_prices["Price"].sum())
age_prices_value.head(10)
```

```python
age_prices_value = age_prices_value.rename(
    columns={"Price": "Total Purchase Value"})
age_prices_value.head(10)
```

```python
age_merged_df = age_grouping_df.merge(age_prices_value, 
                                  on=["Age Ranges"])

age_merged_df.head(10)
```

```python
age_merged_df["Average Purchase Price"] = age_merged_df["Total Purchase Value"] /\
                                                    age_merged_df["Purchase Count"]
age_merged_df["Avg Total Purchase per Person"] = age_merged_df["Total Purchase Value"] /\
                                                    age_merged_df["Players"] 

age_merged_df["Average Purchase Price"] = age_merged_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)

age_merged_df["Avg Total Purchase per Person"] = age_merged_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

age_merged_df["Total Purchase Value"] = age_merged_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)

age_merged_df
```

```python
better_merged = pd.DataFrame(age_merged_df[["Purchase Count",
                                          "Average Purchase Price",
                                          "Total Purchase Value",
                                          "Avg Total Purchase per Person"]])
better_merged
```
